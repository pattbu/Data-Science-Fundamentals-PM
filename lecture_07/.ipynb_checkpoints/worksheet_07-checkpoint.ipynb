### Density-Based Clustering

Follow along with the live coding of the DBScan algorithm.

```python
import numpy as np
from PIL import Image as im
import matplotlib.pyplot as plt
import sklearn.datasets as datasets

#Centers
centers = [[1, 1], [-1, -1], [1, -1]]
X, _ = datasets.make_blobs(n_samples=750, centers=centers, cluster_std=0.4,
                            random_state=0)
plt.scatter(X[:,0],X[:,1],s=10, alpha=0.8)
plt.show()
class DBC():

    def __init__(self, dataset, min_pts, epsilon):
        self.dataset = dataset
        self.min_pts = min_pts
        self.epsilon = epsilon
        self.assignments = [-1 for _ in range(len(self.dataset))]
        self.snaps = []

    def distance(self,i,j):
        #determines distance between point i and j
        return np.linalg.norm(self.dataset[i] - self.dataset[j])
        
    def is_core(self,i):
        #returns true if there is additional "clusters" to be created
        return len(self.get_neighborhood(i))-1 >= self.min_pts
   
    def get_neighborhood(self,i):
        #list of list that stores the neighborhood
        neighborhood = []
        for j in range(len(self.dataset)):
            if i != j and self.distance(i,j) < self.epsilon:
                #If it's not the same point and the point is within distance of epsilon or the radi
                #If conditions are met than the point gets added into the neighborhood
                neighborhood.append(j)
        return neighborhood
    
    def get_unassigned_neighborhood(self,i):
        #Checks if points in the neighborhood of i are assigned to a cluster and returns all points aren't
            neighborhood = self.get_neighborhood(i)
            return [point for point in neighborhood if self.assignments[point] == -1]
    
    def is_unassigned(self, i):
        #Checker for if point i is not associated with a cluster
        return self.assignments[i]==-1
    
    
    def make_cluster(self,i,cluster_num):
        #Creates a clusters around the point i in the cluster given
        self.assignments[i] = cluster_num                    #Assigning the data point to a cluster based on index number
        #self.snapshot(i)
        nbhd_q = self.get_unassigned_neighborhood(i)         #Gathers the data points that are in range of data point i and unassigned
        while nbhd_q:
            next_pt = nbhd_q.pop()                           #The data point FIFO
            if not self.is_unassigned(next_pt):
                #if neighborhood[i] isn't -1 ie the data point is assigned to a cluster
                continue
            self.assignments[next_pt] = cluster_num          #Assigns the data point to the cluster
            #self.snapshot(i)
            if self.is_core(next_pt):
                 nbhd_q += self.get_unassigned_neighborhood(next_pt)
            
    
    def dbscan(self):
        """
        returns a list of assignments. The index of the
        assignment should match the index of the data point
        in the dataset.
        """
        cluster_num = 0
        for i in range(len(self.dataset)):
            if(self.assignments[i] != -1):
               continue
            if self.is_core(i):
                #start building new cluster
                self.make_cluster(i,cluster_num)
                cluster_num += 1
        return self.assignments
    
    def snapshot(self,i):
        #index of the previous data point being looked at
        fig, ax = plt.subplots()
        colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
        colors = np.hstack([colors] * 10)
        plt.scatter(X[:, 0], X[:, 1], color="blue", s=10, alpha=0.8)
        cir = plt.Circle((self.dataset[i][0], self.dataset[i][-1]), radius=0.2, edgecolor='red', facecolor='none')
        ax.add_patch(cir)
        ax.set_aspect('equal') # necessary or else the circles appear to be oval shaped
        plt.close()


"""
dbs.scan: Creates the new clusters
    cluster_num: keeps track of the amount of cluster created
    for loop that goes through the dataset of tuples and se

"""


clustering = DBC(X, 3, .2).dbscan()
colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 10)
plt.scatter(X[:, 0], X[:, 1], color=colors[clustering].tolist(), s=10, alpha=0.8)
plt.show()
```
